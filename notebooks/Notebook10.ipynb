```python
import pandas as pd
```

```python
df = pd.read_excel("reports/report_val.xlsx")
df
```

```python
df["[C] MSC1 (z)"] = ((df["[C] MSC1"] - df["[C] MSC1"].mean()) / df["[C] MSC1"].std())
df["[C] MSC2 (z)"] = ((df["[C] MSC2"] - df["[C] MSC2"].mean()) / df["[C] MSC2"].std())
df = df.rename(columns={"[C] MSC1": "[C] MSC1 (raw)", "[C] MSC2": "[C] MSC2 (raw)"})
df = df.drop(["[C] MSC3", "[C] MSC4"], axis=1)
df
```

```python
df.to_csv("reports/report_val_z.csv")
```

```python
project_rows = []

for db, group_df in df.groupby("DB"):
    project_rows.append({
        "DB": db,
        "[E] LOC": group_df["[E] LOC"].mean(),
        "[C] LCOM1": group_df["[C] LCOM1"].mean(),
        "[C] LCOM2": group_df["[C] LCOM2"].mean(),
        "[C] LCOM3": group_df["[C] LCOM3"].mean(),
        "[C] MSC1 (z)": group_df["[C] MSC1 (z)"].mean(),
        "[C] MSC2 (z)": group_df["[C] MSC2 (z)"].mean(),
    })

project_df = pd.DataFrame.from_records(project_rows)
project_df
```

```python
project_df.to_csv("reports/report_val_projects.csv")
```
